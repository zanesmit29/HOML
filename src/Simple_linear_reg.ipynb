```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/weight-height.csv")
```

```python
df.head(5)
```

```python
plt.plot(df["Height"],df["Weight"], 'o')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Scatter plot weight vs. height')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

lr = LinearRegression()

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
print(f'Train set head: {train_set.head()} \nand train set shape: {train_set.shape}')
```

```python
test_set.shape
```

```python
#Training the model to the training data
lr.fit(train_set[['Height']], train_set["Weight"])
```

```python
print(df["Height"])
```

```python
print(df[["Height"]])
```

```python
test_set["Predicted weight"] = lr.predict(test_set[["Height"]])
```

```python
test_set.head()
```

```python
plt.plot(test_set['Height'], test_set['Predicted weight'], 'o')
plt.xlabel('Weight')
plt.ylabel('Predicted Weight')
plt.title('Actual vs. Predicted')
plt.show()
```
